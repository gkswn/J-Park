```python
df['month'] = df['단속일자'].dt.strftime('%m')
```

```python
df.head()
```

```python
df03 = df[df['month']=='03']
```

```python
df03['단속장소명'].value_counts()
```

```python
where_03 = df03['단속장소명'].value_counts().keys().tolist()
counts_03 = df03['단속장소명'].value_counts().tolist()
df_where03 = pd.DataFrame(where_03)
df_count03 = pd.DataFrame(counts_03)
df_03final = pd.concat([df_where03,df_count03],axis=1)
df_03final.columns = ['단속장소','단속빈도']
df_03final.head(50)
```

```python
df_03final['month'] = '03'
df_03final.head()
```

```python
df_03final.to_excel('3월 단속빈도수.xlsx',encoding='utf-8',index=False)
```
