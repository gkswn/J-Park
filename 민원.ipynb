```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('C:\\Users\\안한주\\Desktop\\전주빅데\\19-21년 주정차단속.xlsx',encoding='utf-8')
df.head()
```

```python
df = df.dropna(axis=0)
```

```python
df.isna().sum()
```

```python
cctv = df[df['단속구분'].str.contains('CCTV')].index
print(cctv)
```

```python
df.drop(cctv,inplace=True)
df.info()
```

```python
df['단속구분'].value_counts()
```

```python
fire = df[df['단속구분'].str.contains('소방서')].index
print(fire)
```

```python
df.drop(fire,inplace=True)
df['단속구분'].value_counts()
```

```python
man_power = df[df['단속구분'].str.contains('인력')].index
print(man_power)
```

```python
df.drop(man_power,inplace=True)
df['단속구분'].value_counts()
```

```python
df.head()
```

```python
df_112 = df[df['단속구분'] =='112민원신고']
```

```python
where_112 = df_112['단속장소명'].value_counts().keys().tolist()
counts_112 = df_112['단속장소명'].value_counts().tolist()
df_where112 = pd.DataFrame(where_112)
df_count112 = pd.DataFrame(counts_112)
df_112final = pd.concat([df_where112,df_count112],axis=1)
df_112final.columns = ['단속장소','단속빈도']
df_112final.head()
```

```python
df_112final.head(50)
```

```python
df_life = df[df['단속구분'] =='생활민원신고']
```

```python
df_life['단속장소명'].value_counts()
```

```python
where_life = df_life['단속장소명'].value_counts().keys().tolist()
counts_life = df_life['단속장소명'].value_counts().tolist()
df_wherelife = pd.DataFrame(where_life)
df_countlife = pd.DataFrame(counts_life)
df_lifefinal = pd.concat([df_wherelife,df_countlife],axis=1)
df_lifefinal.columns = ['단속장소','단속빈도']
df_lifefinal.head(50)
```
